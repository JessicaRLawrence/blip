```python
from __future__ import division
import numpy as np
import matplotlib
matplotlib.rcParams['figure.figsize'] = (14,10)
import matplotlib.pyplot as plt
```

```python
times = np.arange(0, 1e6, 10 )
alpha = 2*np.pi*times/(60*60*24*365.5)


beta = np.array([ 0, 1, 2])*2*np.pi/3


# distance to earth from sun in meters
a = 149.6*1e9
Larm = 2.5e9


# eccentricity
e = Larm/(2*a*np.sqrt(3))

## For 1
x1 = a*np.cos(alpha) + a*e*(np.sin(alpha)*np.cos(alpha)*np.sin(beta[0]) - (1 + (np.sin(alpha))**2 )* np.cos(beta[0]))
y1 = a*np.sin(alpha) + a*e*(np.sin(alpha)*np.cos(alpha)*np.cos(beta[0]) - (1 + (np.cos(alpha))**2 )* np.sin(beta[0]))
z1 = - np.sqrt(3)*a*e*np.cos(alpha - beta[0])

## For 2
x2 = a*np.cos(alpha) + a*e*(np.sin(alpha)*np.cos(alpha)*np.sin(beta[1]) - (1 + (np.sin(alpha))**2 )* np.cos(beta[1]))
y2 = a*np.sin(alpha) + a*e*(np.sin(alpha)*np.cos(alpha)*np.cos(beta[1]) - (1 + (np.cos(alpha))**2 )* np.sin(beta[1]))
z2 = - np.sqrt(3)*a*e*np.cos(alpha - beta[1])

## For 3
## For 2
x3 = a*np.cos(alpha) + a*e*(np.sin(alpha)*np.cos(alpha)*np.sin(beta[2]) - (1 + (np.sin(alpha))**2 )* np.cos(beta[2]))
y3 = a*np.sin(alpha) + a*e*(np.sin(alpha)*np.cos(alpha)*np.cos(beta[2]) - (1 + (np.cos(alpha))**2 )* np.sin(beta[2]))
z3 = - np.sqrt(3)*a*e*np.cos(alpha - beta[2])

## distance b/w the satellites
r12 = np.sqrt( (x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)
r32 = np.sqrt( (x2-x3)**2 + (y2-y3)**2 + (z2-z3)**2)
r31 = np.sqrt( (x1-x3)**2 + (y1-y3)**2 + (z1-z3)**2)
```

```python
plt.semilogx(times, r12, label='1-2')
plt.semilogx(times, r32, label='3-2')
plt.semilogx(times, r31, label='1-3')
plt.axhline(Larm,linestyle='--' ,label='reference' )
plt.legend()
plt.ylim(2.501e9, 2.499e9)
```

```python
plt.semilogy(times, np.abs((r12-r32)/r12), label='12-32')
plt.semilogy(times, np.abs((r12-r31)/r12), label='12-31')
plt.semilogy(times, np.abs((r32-r31)/r32), label='32-31')
plt.show()
```
